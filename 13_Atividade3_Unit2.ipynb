```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando os dados fictícios
dados_vendas = {
    "Dia": ["Seg", "Ter", "Qua", "Qui", "Sex", "Sáb", "Dom"],
    "Vendas": [2500, 2700, 3000, 3200, 3500, 4000, 4500],
    "Clientes": [200, 210, 220, 230, 250, 300, 320],
    "Lucro": [500, 550, 600, 650, 700, 800, 900]
}

df_vendas = pd.DataFrame(dados_vendas)

# Removendo a coluna "Dia" para calcular a correlação
df_corr = df_vendas.drop(columns=["Dia"])

# Criando Heatmap da correlação
plt.figure(figsize=(8,5))
sns.heatmap(df_corr.corr(), annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Matriz de Correlação entre as Variáveis")
plt.show()

```

```python
# Criando Gráfico de Barras - Total de Vendas por Dia

plt.figure(figsize=(8, 5))
plt.bar(df_vendas["Dia"], df_vendas["Vendas"], color="royalblue", alpha=0.7)
plt.title("Total de Vendas por Dia")
plt.xlabel("Dia da Semana")
plt.ylabel("Total de Vendas")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Criando Gráfico de Dispersão - Clientes x Vendas

plt.figure(figsize=(8, 5))
plt.scatter(df_vendas["Clientes"], df_vendas["Vendas"], color="red", alpha=0.7)
plt.title("Relação entre Número de Clientes e Vendas")
plt.xlabel("Número de Clientes")
plt.ylabel("Total de Vendas")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
