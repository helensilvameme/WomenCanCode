```python
import matplotlib.pyplot as plt

# Definindo os horários de 0h a 24h
horarios = list(range(25))

# Criando os valores da temperatura: sobe até 30°C às 12h e desce até 18°C às 24h
temperaturas = [15 + (i * 1.25) if i <= 12 else 30 - ((i - 12) * 1.0) for i in horarios]

# Criando o gráfico
plt.figure(figsize=(10,5))
plt.plot(horarios, temperaturas, marker='o', linestyle='-', color='b', label="Temperatura (°C)")

# Personalizando o gráfico
plt.title("Evolução da Temperatura Durante o Dia")
plt.xlabel("Horário (h)")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.legend()

# Exibindo o gráfico
plt.show()
```
